#Step1: check data info, drop bad columns


```python
census = pd.read_csv('drive/My Drive/bubble_tea/neighbourhood-profiles-2016-csv.csv').copy()
```

```python
census.info()
```

```python
census.columns
```

drop all columns with names starts with 'Unnamed'

```python
droplist = [col for col in census.columns if col.startswith('Unnamed')]
census.drop(droplist, axis=1, inplace=True)
census.drop(columns=['Category','Data Source','City of Toronto','_id'], inplace=True)
census.drop([1], inplace=True)
census.head()
```

#Step2: Check Unique data types and convert objects to float


Check unique data types


```python
uniq_types =[]
def check_types(x):
  current_type = type(x)
  if current_type not in uniq_types:
    uniq_types.append(type(x))
    
census_copy = census.copy()
columns= census.columns

for col in columns:
  if col == 'Topic':
    continue
  else:
    census_copy[col].apply(lambda x : check_types(x))
#This case we only have str and float. 

```

Removing extra symbols. Convert strings to pretty numbers

```python

print(columns)
hood_name= census_copy.iloc[:,2:].columns


for col in hood_name:
  census_copy[col]= census_copy[col].str.replace(',','').str.replace('%','')
  census_copy[col] = census_copy[col].astype('float')  #returned value requires a variable to hold it

census_copy.head()
#census_copy.dtypes >> return only float type
```

#Step 3: create new df with selected columns and reshape it


```python
census_copy['Topic'].unique()
topics = census_copy['Topic'].unique().tolist()  #array to list

```

Manually drop off totally nonsense/repetitive columns
select columns to keep, then create new df with selected columns


```python
topics_selected = ['Population and dwellings',
 'Age characteristics','Marital status','Family characteristics of adults','Income of households in 2015','Income of individuals in 2015', 'Ethnic origin population','Highest certificate, diploma or degree','Household characteristics',
 'Core housing need','Highest certificate, diploma or degree','Work activity during the reference year', 'Occupation - National Occupational Classification (NOC) 2016',
 ]
len(topics_selected)
```

```python
filt =census_copy['Topic'].isin(topics_selected)   #series.isin return true/false
census_selected = census_copy[filt]
census_selected.drop(columns=['Topic'],inplace=True)
census_selected.reset_index(drop=True, inplace=True)
census_selected.head()

```

#Step 4: Transpose table to prepare for future merge with others


```python
transposed = census_selected.T
print(transposed.columns)
print(transposed.head())

```

Use first row as header

```python
new_header = transposed.iloc[0] #grab the first row for the header
 #take the data less the header row
transposed.columns = new_header #set the header row as the df header
transposed = transposed[1:]
transposed.head(20)
transposed.to_csv(f'{bbt_folder}transposed.csv')
```
